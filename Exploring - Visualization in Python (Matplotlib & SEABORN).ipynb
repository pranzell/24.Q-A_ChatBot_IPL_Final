## Line Curve

```python
%matplotlib inline
```

```python
x = np.linspace(1,100,5)
```

```python
y = x**2
```

```python
plt.plot(x,y)
```

## Multiple Line Curves

```python
plt.subplot(1,4,1)
plt.plot(x,y)

plt.subplot(1,4,2)
plt.plot(x,y)

plt.subplot(1,4,3)
plt.plot(x,y)

plt.subplot(1,4,4)
plt.plot(x,y)
```

## In-Display Line Curves

```python
fig = plt.figure()

axes1=fig.add_axes([0.0, 0.0, 0.9, 0.9])
axes2=fig.add_axes([0.1, 0.4, 0.4, 0.4])
axes1.plot(x,y)
axes2.plot(y,x)
```

## 2-System Line Curves 

```python
fig = plt.figure()

axes=fig.add_axes([0,0,1,1])

axes.plot(x,x**2, label='X d', c='r')
axes.plot(x,x**2.4, label='Y e', c='b')
```

## Line Curves with markings

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y, ls='-', lw=1, c='GREEN', marker='o', markersize=20)
```

## Scatter Plot

```python
plt.scatter(x,y)
```

## Histogram

```python
plt.hist(y)
```
